```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("StudentsPerformance.csv")

df.head()
```

```python
print("First 10 Rows")
display(df.head(10))

print("Last 10 Rows")
display(df.tail(10))
```

```python
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nInfo:")
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.loc[5, "math score"] = np.nan
df.loc[15, "reading score"] = np.nan
df.loc[25, "writing score"] = np.nan

df.isnull().sum()
```

```python
df["math score"] = df["math score"].fillna(df["math score"].mean())

df["reading score"] = df["reading score"].fillna(df["reading score"].mean())

df["writing score"] = df["writing score"].fillna(df["writing score"].mean())

df.isnull().sum()
```

```python
print("Duplicate Rows:", df.duplicated().sum())

df = df.drop_duplicates()
```

```python
print("Gender")
print(df["gender"].unique())

print("\nLunch")
print(df["lunch"].unique())

print("\nTest Preparation")
print(df["test preparation course"].unique())

print("\nRace")
print(df["race/ethnicity"].unique())
```

```python
df["Total Score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
)

df.head()
```

```python
df["Average Score"] = df["Total Score"] / 3

df.head()
```

```python
df["Percentage"] = (df["Total Score"] / 300) * 100

df.head()
```

```python
def grade(percent):
    if percent >= 90:
        return "A+"
    elif percent >= 80:
        return "A"
    elif percent >= 70:
        return "B"
    elif percent >= 60:
        return "C"
    else:
        return "D"

df["Grade"] = df["Percentage"].apply(grade)

df.head()
```

```python
df["Result"] = np.where(
    (df["math score"] >= 35) &
    (df["reading score"] >= 35) &
    (df["writing score"] >= 35),
    "Pass",
    "Fail"
)

df.head()
```

```python
df.insert(0, "Student_ID", range(1, len(df) + 1))

df.head()
```

```python
print("Total Students:", len(df))

print("\nGender Count")
print(df["gender"].value_counts())

print("\nAverage Math Score:", round(df["math score"].mean(), 2))
print("Average Reading Score:", round(df["reading score"].mean(), 2))
print("Average Writing Score:", round(df["writing score"].mean(), 2))
```

```python
attendance = pd.DataFrame({
    "Student_ID": df["Student_ID"],
    "Attendance": np.random.randint(60, 101, len(df))
})

attendance.head()
```

```python
attendance.to_csv("student_attendance.csv", index=False)

print("Attendance dataset saved.")
```

```python
merged = pd.merge(
    df,
    attendance,
    on="Student_ID",
    how="left"
)

merged.head()
```

```python
merged.groupby("Grade")["Attendance"].mean().round(2)
```

```python
merged[merged["Attendance"] < 75]
```

```python
merged.to_csv("cleaned_students.csv", index=False)

print("Cleaned dataset saved successfully.")
```

```python
print("=" * 50)
print("STUDENT PERFORMANCE REPORT")
print("=" * 50)

print(f"Total Students : {len(merged)}")

print(f"Male Students : {(merged['gender']=='male').sum()}")
print(f"Female Students : {(merged['gender']=='female').sum()}")

print(f"\nAverage Math Score : {merged['math score'].mean():.2f}")
print(f"Average Reading Score : {merged['reading score'].mean():.2f}")
print(f"Average Writing Score : {merged['writing score'].mean():.2f}")

print(f"\nHighest Percentage : {merged['Percentage'].max():.2f}")
print(f"Lowest Percentage : {merged['Percentage'].min():.2f}")

print(f"\nPass Students : {(merged['Result']=='Pass').sum()}")
print(f"Fail Students : {(merged['Result']=='Fail').sum()}")

print("\nGrade Distribution")
print(merged["Grade"].value_counts())

print("=" * 50)
```
